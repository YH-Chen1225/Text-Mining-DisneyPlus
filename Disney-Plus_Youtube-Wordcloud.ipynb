```python
# Importing all needed libraries
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
# Reading the CSV file with the YouTube Comments
df = pd.read_csv(r"YT.csv", encoding ="latin-1")
```

```python
# Importing the stopwords list not to show them in the WordCloud
comment_words = ' '
stopwords = set(STOPWORDS)
```

```python
# Iterate through the CSV file 
for val in df.Comments: 
      
    # Typecaste each val to string 
    val = str(val) 
  
    # Split the value 
    tokens = val.split() 
      
    # Convert each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
          
    for words in tokens: 
        comment_words = comment_words + words + ' '
```

```python
# Setting up the WordCloud
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10, 
                colormap='gist_heat').generate(comment_words)
```

```python
# Plotting the WordCloud image 

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show()
```
